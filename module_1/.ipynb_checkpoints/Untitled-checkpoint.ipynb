```python
#Вопрос 22.Сколько суммарно вышло фильмов летом (за июнь, июль, август)?
import pandas as pd
data = pd.read_csv('movie_bd_v5.csv')
df22 = data[['imdb_id', 'release_date']].copy().set_index('imdb_id')
df22.release_date = df22['release_date'].str.split('/', expand=True)
df22.rename(columns={"release_date": "month"}, inplace=True)
df5 = df22.loc[df22['month'].isin(['6', '7', '8'])]
df5.month.count()


#print(pd.pivot_table(df4, index='imdb_id', aggfunc='count').nlargest(1, 'month'))
```

```python
#Вопрос 23. Для какого режиссера зима — самое продуктивное время года?
import pandas as pd
data = pd.read_csv('movie_bd_v5.csv')
df2 = data[['imdb_id', 'release_date', 'director']].copy().set_index('imdb_id')
df2.director = df2.director.str.split('|')
df3 = df2.explode('director')
df3.release_date = df3['release_date'].str.split('/', expand=True)
df3.rename(columns={"release_date": "month"}, inplace=True)
df4 = df3.loc[df3['month'].isin(['12', '1', '2'])]
print(pd.pivot_table(df4, index='director', aggfunc='count').nlargest(1, 'month'))
```

```python
#Вопрос 24. Какая студия даёт самые длинные названия своим фильмам по количеству символов?

import pandas as pd
data = pd.read_csv('movie_bd_v5.csv')
data_24 = data.copy()
data_24['new_title'] = data_24['original_title'].str.len()
data_24.production_companies = data_24.production_companies.str.split('|')
data_24 = data_24.explode('production_companies')
data_24.groupby(by='production_companies')['new_title'].mean().sort_values(ascending=False)
# Способ 2
#print(pd.pivot_table(data_24, values='new_title', index='production_companies', aggfunc='mean').nlargest(5, 'new_title'))

```

```python
#Вопрос 25. Описания фильмов какой студии в среднем самые длинные по количеству слов?
import pandas as pd
data = pd.read_csv('movie_bd_v5.csv')
data_25 = data.copy()
data_25['new_overview'] = data_25['overview'].str.split().str.len()
data_25.production_companies = data_25.production_companies.str.split('|')
data_25 = data_25.explode('production_companies')
#data_24.groupby(by='production_companies')['new_title'].mean().sort_values(ascending=False)
# Способ 2
print(pd.pivot_table(data_25, values='new_overview', index='production_companies', aggfunc='mean').nlargest(15, 'new_overview'))
```

```python
#26
import pandas as pd
import numpy as np
data26 = data.copy()
procent_99 = np.percentile(data26.vote_average, 99)
# Записываем в переменную оценку равную 99% от средней
data26[data26.vote_average >= procent_99].sort_values(
    by='vote_average', ascending=False)['original_title']
# Выводим все фильмы с оценкой выше или равной 99%
```
